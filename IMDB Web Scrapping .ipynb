```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from time import sleep
from time import time 
from IPython.display import clear_output


name = []
IMDB_rating = []
year_of_release = []
metascore = []
number_of_votes = []


request = 0
start_time = time()

years= [str(i) for i in range(2000,2018)]
pages= [str(i) for i in range(1,5)]


for year in years:
    
    for page in pages:
        
        response = requests.get('http://www.imdb.com/search/title?release_date={}&sort=num_votes,desc&page={}'.format(year,page))
        
        sleep(np.random.randint(8,15))
        
        request +=1 
        lapsed_time= time()- start_time
        print("Request : {}, Frequency : {}".format(request, request/lapsed_time))
        clear_output(wait= True)

        
        html_soup= BeautifulSoup(response.text, "html.parser")
        movie_containers= html_soup.find_all('div', {'class': 'lister-item mode-advanced'})

        for i in movie_containers:
            if i.find('span', {"class": "metascore favorable"}) is not None:
                name.append(i.h3.a.text)
                year_of_release.append(i.h3.find('span', {"class": "lister-item-year text-muted unbold"}).text)
                IMDB_rating.append(float(i.strong.text))
                metascore.append(int(i.find('span', {"class": "metascore favorable"}).text))
                number_of_votes.append(i.find("p", {"class": "sort-num_votes-visible"}).find("span", {"name":"nv"}).text)
            else:
                pass

df= pd.DataFrame({"Movie_name": name,
                    "IMDB_rating": IMDB_rating,
                    "year_of_release": year_of_release,
                    "metascore": metascore,
                    "number_of_votes": number_of_votes})


df["year"]= df["year_of_release"].str[-5:-1].astype(int)
df = df.drop(['year_of_release'],1)
df.to_csv(" IMDb movie data.csv")
```
